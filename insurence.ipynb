```python
import pandas as pd
import numpy as np

def clean_date(x):
    
    if pd.isna(x):
        return np.nan
    
    # Excel serial date
    if isinstance(x, (int, float)):
        return pd.to_datetime(x, unit='D', origin='1899-12-30')
    
    x = str(x)
    
    # YYYY_MM_DD
    if '_' in x:
        return pd.to_datetime(x, format='%Y_%m_%d', errors='coerce')
    
    # DD/MM/YYYY
    if '/' in x:
        return pd.to_datetime(x, dayfirst=True, errors='coerce')
    
    # DD-Mon-YYYY
    return pd.to_datetime(x, errors='coerce')


df['Policy_Start_Date'] = df['Policy_Start_Date'].apply(clean_date)
```
